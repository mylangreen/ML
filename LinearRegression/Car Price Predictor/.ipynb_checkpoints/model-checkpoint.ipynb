### Download The Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

```python
car_df = pd.read_csv("hf://datasets/VarunKumarGupta2003/Car-Price-Dataset/car_sales_data.csv")
```

```python
car_df.head()
```

### Inspect The Data

```python
car_df.info()
```

### Visualize The Data

### 1. Distribution

#### a.) Year

```python
car_df['Year'].describe()
```

```python
fig = px.histogram(
    car_df,
    x='Year',
    marginal='box',
    nbins=21,
    title="Year Distribution"
)
fig.update_layout(bargap=0.1, title={"y":0.9, "x":0.5})
```

#### b.) Engine Size

```python
car_df['Engine Size'].describe()
```

```python
fig = px.histogram(
    car_df,
    x='Engine Size',
    nbins=5,
    marginal='box',
    title="Engine Size Distribution"
)
fig.update_layout(bargap=0.1, title={ "y":0.9, "x":0.5})
```

#### c.) Mileage

```python
car_df['Mileage'].describe()
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(data=car_df, x='Mileage', bins=20, shrink=0.9, alpha=0.2, kde=True)
plt.title('Mileage Distribution')
plt.show()
```

```python
fig = px.histogram(
    car_df,
    x="Mileage",
    nbins=20,
    marginal='box',
    width=1000,
    height=500,
    title='Mileage Distribution'
)
fig.update_layout( bargap=0.1, title={ 'y':0.9, 'x':0.5} )
```

### d.) Price Distribution

```python
car_df['Price'].describe()
```

```python
fig = px.histogram(
    car_df,
    x="Price",
    marginal="box",
    title="Price Distribution",
    nbins=40,
    width=1000,
    height=500
)
fig.update_layout(bargap=0.1, title={"y":0.9, "x":0.5})
```

### e.) Make

```python
car_df['Make'].describe()
```

```python
sns.countplot(data=car_df, x='Make')
plt.title('Distribution Of Car Make')
plt.show()
```

```python
fig = px.histogram(
    car_df,
    x='Make',
    title='Distribution Of Car Make'
)
fig.update_layout(bargap=0.1)
```

### f.) Transmission

```python
fig = px.histogram(
    car_df,
    x='Transmission',
    title='Distribution Of Car Transmission'
)
fig.update_layout(bargap=0.1)
```

### h.) Model

```python
fig = px.histogram(
    car_df,
    x='Model',
    title='Distribution Of Car Model'
)
fig.update_layout(bargap=0.1)
```

### i.) Fuel Type

```python
fig = px.histogram(
    car_df,
    x='Fuel Type',
    title='Distribution Of Fuel Type'
)
fig.update_layout(bargap=0.1)
```

## 2. Relationships With The Target

#### a.) Year

```python
plt.figure(figsize=(10,5))
sns.scatterplot(
    data=car_df,
    x='Year',
    y='Price',
    alpha=0.5
)
plt.title('Year vs Price')
plt.show()
```
